```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Files to Load
school_data = "schools.csv"
student_data = "students.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.read_csv(school_data)
student_df = pd.read_csv(student_data)

# Show School dataframe to validate
school_df.head()
```

```python
# Show Students dataframe to validate
student_df.head()
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])

# Show dataframe to validate merge
school_data_complete.head()
```

```python
# Start with DISTRICT SUMMARY

# Calculate the total number of schools using nunique
school_count=school_data_complete["school_name"].nunique()
#Validate
print(school_count)
```

```python
#Calculate the total number of students -- NOTE: not unique as there can be students with same name in different school
student_count=school_data_complete["student_name"].count()
#Validate
print(student_count)
```

```python
#Calculate the total budget
total_budget=school_df["budget"].sum()
#Validate
print(total_budget)
```

```python
#Calculate the average math score
math_average=student_df["math_score"].mean()
#Validate
print(math_average)
```

```python
#Calculate the average reading score
reading_average=student_df["reading_score"].mean()
print(reading_average)
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
average_overall_score=(math_average+reading_average)/2
#Validate
print(average_overall_score)
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
math_passing = school_data_complete[school_data_complete["math_score"] >= 70].count()["school_name"]
math_passing_percent = (math_passing / student_count) * 100

#Validate
print(math_passing_percent)
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
reading_passing = school_data_complete[school_data_complete["reading_score"] >= 70].count()["school_name"]
reading_passing_percent = (reading_passing / student_count) * 100

#Validate
print(reading_passing_percent)
```

```python
#Create a dataframe to hold the above results
District_Summary = pd.DataFrame({"Total Schools"               :[school_count], 
                                 "Total Students"              :[student_count],
                                 "Total Budget"                :[total_budget], 
                                 "Average Math Score"          :[math_average],
                                 "Average Reading Score"       :[reading_average], 
                                 "% Passing Math"              :math_passing_percent,
                                 "% Passing Reading"           :reading_passing_percent, 
                                 "Overall Average Score"       :average_overall_score})
```

```python
# Set index and display dataframe
District_Summary = District_Summary[["Total Schools", "Total Students", "Total Budget", "Average Math Score",
                                 "Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Average Score"]]
District_Summary
```

```python
#School Type
school_types = school_df.set_index(["school_name"])["type"]

#validate
print(school_types)
```

```python
#Total Students
per_school_count=school_df.set_index(["school_name"])["size"]

#validate
print(per_school_count)
```

```python
#Total School Budget
school_budgets=school_df.set_index(["school_name"])["budget"]

#Validate
print(school_budgets)
```

```python
#Per Student Budget
per_student_budget=school_budgets/per_school_count

#Validate
print(per_student_budget)
```

```python
#Average Math Score
average_math_per_school = school_data_complete.groupby(["school_name"]).mean()["math_score"]

#validate
print(average_math_per_school)
```

```python
#Average Reading Score
average_reading_per_school = school_data_complete.groupby(["school_name"]).mean()["reading_score"]

#Validate
print(average_reading_per_school)
```

```python
#% Passing Math
students_passing_math =  school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name").count()["Student ID"]
percent_students_passing_math=(students_passing_math/per_school_count)*100
#Validate
print(percent_students_passing_math)
```

```python
#% Passing Reading
students_passing_reading =  school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name").count()["Student ID"]
percent_students_passing_reading=(students_passing_reading/per_school_count)*100
#Validate
print(percent_students_passing_reading)
```

```python
#Overall Passing Rate (Average of the above two)
overall_passing_rate=(percent_students_passing_reading+percent_students_passing_math)/2

#Validate
print(overall_passing_rate)
```

```python
#Create a dataframe to hold the above results
School_Summary = pd.DataFrame({"School Type"             : school_types, 
                               "Total Students"          : per_school_count, 
                               "Total School Budget"     : school_budgets, 
                               "Per Student Budget"      : per_student_budget,
                               "Average Math Score"      : average_math_per_school, 
                               "Average Reading Score"   : average_reading_per_school,
                               "% Passing Math"          : percent_students_passing_math, 
                               "% Passing Reading"       : percent_students_passing_reading,
                               "Overall Passing Rate"    : overall_passing_rate})  
```

```python
# Set index 
School_Summary = School_Summary[["School Type","Total Students","Total School Budget","Per Student Budget",
                                "Average Math Score","Average Reading Score","% Passing Math","% Passing Reading",
                                 "Overall Passing Rate"]]
School_Summary
```

```python
#START PART 3
#Top Performing Schools (By Passing Rate)

#Sort and display the top five schools in overall passing rate
Top_Schools = School_Summary.sort_values("Overall Passing Rate", ascending=False)

Top_Schools.head(5)
```

```python
#START PART FIVE
#Math Scores by Grade

#Math Scores in a table
ninth_grade_score =  school_data_complete[school_data_complete["grade"] == "9th"].groupby("school_name").mean()["math_score"]
tenth_grade_score =  school_data_complete[school_data_complete["grade"] == "10th"].groupby("school_name").mean()["math_score"]
eleventh_grade_score =  school_data_complete[school_data_complete["grade"] == "11th"].groupby("school_name").mean()["math_score"]
twelfth_grade_score =  school_data_complete[school_data_complete["grade"] == "12th"].groupby("school_name").mean()["math_score"]

Math_Summary = pd.DataFrame({"9th Grade"             : ninth_grade_score, 
                             "10th Grade"            : tenth_grade_score, 
                             "11th Grade"            : eleventh_grade_score,
                             "12th Grade"            : twelfth_grade_score})  
# Set index 
Math_Summary = Math_Summary[["9th Grade","10th Grade","11th Grade","12th Grade"]]

Math_Summary
```

```python
#START PART 6
#Reading Score by Grade
#Perform the same operations as above for reading scores

#Reading Scores in a table
r_ninth_grade_score =  school_data_complete[school_data_complete["grade"] == "9th"].groupby("school_name").mean()["reading_score"]
r_tenth_grade_score =  school_data_complete[school_data_complete["grade"] == "10th"].groupby("school_name").mean()["reading_score"]
r_eleventh_grade_score =  school_data_complete[school_data_complete["grade"] == "11th"].groupby("school_name").mean()["reading_score"]
r_twelfth_grade_score =  school_data_complete[school_data_complete["grade"] == "12th"].groupby("school_name").mean()["reading_score"]

Reading_Summary = pd.DataFrame({"9th Grade"             : r_ninth_grade_score, 
                             "10th Grade"            : r_tenth_grade_score, 
                             "11th Grade"            : r_eleventh_grade_score,
                             "12th Grade"            : r_twelfth_grade_score})  
# Set index 
Reading_Summary = Reading_Summary[["9th Grade","10th Grade","11th Grade","12th Grade"]]

Reading_Summary
```

```python
#START PART 7
#Scores by School Spending
#Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:

#Set bins
spending_bins= [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

#Validate
pd.cut(School_Summary["Per Student Budget"], spending_bins, labels=group_names)
```

```python
# Display our School Summary data frame into our bins
School_Summary["Spending Ranges"] = pd.cut(per_student_budget, spending_bins, labels=group_names)

# Indicators for each Spending Ranges 
spending_math_score = School_Summary.groupby(["Spending Ranges"]).mean()['Average Math Score']
spending_reading_score = School_Summary.groupby(["Spending Ranges"]).mean()['Average Reading Score']
spending_passing_math = School_Summary.groupby(["Spending Ranges"]).mean()['% Passing Math']
spending_passing_reading = School_Summary.groupby(["Spending Ranges"]).mean()['% Passing Reading']
spending_overall_passing_rate = (spending_passing_math + spending_passing_reading) / 2
```
